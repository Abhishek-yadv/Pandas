## Series:- 
                A Series is a one-dimensional labeled array. It is similar to a NumPy array but has additional capabilities. A Series consists of two arrays: one for the data and another for the labels (index). The data can be of any data type, and the labels must be unique. You can think of a Series as a labeled list.

```python
mydata = ['Boat', 'Car', 'Bike', 'Truck']
myserios1 = pd.Series(mydata)
```

```python
print(myseries1)
```

```python
mydata = [1, 55, 99, 43]
myseries2 = pd.Series(mydata)
print(myseries2)
```

```python
import pandas as pd

data = [10, 20, 30, 40, 50]
index = ['A', 'B', 'C', 'D', 'E']

series = pd.Series(data, index=index)
print(series)
```

## DataFrame:- 
                        A DataFrame is a two-dimensional labeled data structure, similar to a spreadsheet or a SQL table. It is the primary data structure used in pandas for most data manipulations. A DataFrame consists of rows and columns, where each column can contain data of different types. You can think of a DataFrame as a collection of Series that share the same index.

```python
mydfdata = [('Boat', 1), ('Car', 55), ('Bike',99), ('Truck', 43)]

mydf = pd.DataFrame(mydfdata)

mydf
```

```python
mydfdata = [('Boat', 1), ('Car', 55), ('Bike',99), ('Truck', 43)]

mydf = pd.DataFrame(mydfdata, columns=['thing', 'count'])

mydf
```

```python
mydf = pd.DataFrame(mydfdata, columns=['thing', 'count'])

mydf
```
